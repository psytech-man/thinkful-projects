```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#matplotlib inline
```

```python
normalvar = np.random.normal(10, 2, 100)
plt.hist(normalvar)
plt.axvline(normalvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normalvar.mean() + normalvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normalvar.mean() - normalvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

Both the mean and std are valid descriptors by definition because this is a normal distribution.

```python
paretovar = np.random.pareto(10, 100)
plt.hist(paretovar)
plt.axvline(paretovar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(paretovar.mean() + paretovar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(paretovar.mean() - paretovar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean and std are not valid descriptors because this is not a normal distribution. The pareto distribution looks like an exponential curve where most data points cluster between -1 std and the mean on this graph. By definition, only 34% of the data should cluster there, so we know that this distribution is not normal.

```python
gammavar = np.random.gamma(50, 1, 100)
plt.hist(gammavar)
plt.axvline(gammavar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(gammavar.mean() + gammavar.std(), color='r', linestyle='dashed')
plt.axvline(gammavar.mean() - gammavar.std(), color='r', linestyle='dashed')
plt.show()
```

Both the mean and std appear to be valid descriptors for the gamma distribution. The mean line is not exactly at 50, however, and the tails appear a bit too big, especially the lower tail that is more than 1 std below the mean. That being said, I believe the mean and std still prove useful info.

```python
binomialvar = np.random.binomial(666, 1/3, 100)
plt.hist(binomialvar)
plt.axvline(binomialvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomialvar.mean() + binomialvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomialvar.mean() - binomialvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean looks like a valid descriptor for this binomial distribution. I set the probability as 1/3 and the number of experiments as 666, so I would expect the mean to be 666 * 1/3, which is 222. That is where the mean line was drawn. The std does not look like a valid descriptor. I say this because there appear to be many more data points between -1 std and the mean compared to the range between the mean and +1 std. For a normal distribution, one would expect to see both ranges capturing ~34% of the data points, which is not what I see here.

```python
exponentialvar = np.random.exponential(.5, 100)
plt.hist(exponentialvar)
plt.axvline(exponentialvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(exponentialvar.mean() + exponentialvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(exponentialvar.mean() - exponentialvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean and std are not valid descriptors for the exponential distribution. The explanation is idential to the one I gave for the pareto distribution, which looks very similar when plotted on the histogram.

```python
chisquarevar = np.random.chisquare(3, 100)
plt.hist(chisquarevar)
plt.axvline(chisquarevar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(chisquarevar.mean() + chisquarevar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(chisquarevar.mean() - chisquarevar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean and std are not valid descriptors for the chisquare distribution. The reasoning is almost idential to that which was given for the pareto distribution. There is one additional problem here. There are two significant outliers, one of which is a very extreme outlier. The presence of these outliers will skew the mean in an unhelpful way.

```python
norm1 = np.random.normal(5, .5, 100)
norm2 = np.random.normal(10, 1, 100)
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(norm3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

The mean looks to be a useful descriptor. The data looks roughly normally distributed and without outliers. That being said, there appears to be more variance and "noise" in the data that lies outside 1 std of the mean. The bins at ~13 and ~17 are unusually large. I suspect that this compromises the usefulness of the std descriptor.

Reflecting upon all the distributions given above, I notice that all the distributions that cannot be described effectively by the mean and std are not normal distributions. They do not look like bell curves. They are not symmetrical. They also have outliers and uneven tails.
