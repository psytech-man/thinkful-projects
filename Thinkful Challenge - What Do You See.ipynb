```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame()
df = pd.read_csv('survey.csv')
```

```python
clean_list = df['Age'].sort_values().tolist()
clean_list = clean_list[6:(len(clean_list)-2)]
age_df = df['Age'].value_counts().sort_index()
age_df2 = age_df[6:(len(age_df)-2)]
```

The two figures below represent the distribution of ages of survey respondents in different ways. The box plot on the left shows the interquartile range and the range of outliers, up to 1.5x the interquartile range. This specific box plot indictates that the interquartile range of respondent age is roughly 28 to 35. The lower bound of the box has a y-axis value that is in the 25th percentile for the total age range and the lower bound of the box has a y-axis value that is in the 75th percentile. The "whiskers" of the plot show the range of values that are up at 1.5x above or below the interquartile range. Finally, the circles above the upper whisker represent the outlier data points that are more than 1.5x above the interquartile range. The large number of circles indicates that there are many respondents to this survey who are much older than most of the respondents. This indicates variability in the age data.

The bar graph to the right gives a clearer view of the distribution of ages. The x-axis shows age values and the y-axis shows the frequency those ages in the survey. Thus, each bar tells you how many survey respondents marked that value as their age. This graph shows you the range of values (a few years below 20 and a few years above 70) and the general shape of the distribution. This distribution would appear to be roughly normally distributed if you exluded all respondents older than 45. It also suggests that the mean age is somewhere close to 30.

```python
plt.figure(figsize=(18, 5))

plt.subplot(1, 2, 1)
plt.boxplot(clean_list, notch=True)
plt.ylabel('Ages')
plt.title('Box Plot for Respondent Ages')

plt.subplot(1, 2, 2)
plt.bar(left=age_df2.keys(), height=age_df2.values)
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.title('Distribution of Respondent Ages')
plt.show()
```

The pie chart below shows what percentage of respondents indicated that they have sought treatment for a mental health condition. Although the percentages are given inside the pie for sake of clarity, a quick glance at the visual immediately shows you there was a 50/50 split - about half of respondents have sought treatment and about half of respondents have not.

```python
plt.figure(figsize=(6, 6))
plt.pie(df['treatment'].value_counts(), colors=['g', 'y'], labels=['Yes', 'No'], labeldistance=1.12, autopct= '%1.1f%%', shadow=True)
plt.title('Have you sought treatment for a mental health condition?')
plt.show()
```
